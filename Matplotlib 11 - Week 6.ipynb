#### Sample Data Set
 
Kali ini kita akan menggunakan sample data et yang dihasilkan dari proses pengacakan. Kita akan menggunakan method `np.random.normal` dimana method ini akan menghasilkan seekelompok bilangan acak ini akan membentuk distribusi normal. 

```python
np.random.seed(2)

# loc mendefinisikan titik tengah / mean.
# scale mengatur sebaran data (standar deviasi).
# size menentukan jumlag data point yang dihasilkan.
data = np.random.normal(loc=100, scale=10, size=200)
data
```

#### Simple Violin Plot

Untuk membuat simple violin plot dari sebuah data set, kita bisa memanfaatkan method `plt.violinplot` dan menyertakan data set pembentuknya. Dikarenakan ini merupakan distribusi normal, maka nilai mean dan mediannya berada di posisi yang sama.

```python
plt.violinplot(data)

# set label title, sumbu x, dan sumbu y.
plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')


plt.show()
```

#### Pengaturan pada Violin Plot

Dalam membuat violin plot, kita bisa melakukan pengaturan dengan menyertakan parameter dalam method `plt.violinplot` yaitu:
- `showextrema` digunakan untuk menampilkan garis tengah dan garis yang mengindikasikan nilai max dan nilai min dari data set.
- `showmeans` dan `showmedians` digunakan untuk menunjukkan indikator nilai mean dan median.
- `quantiles` akan menerima nilai berupa collection data types. Kita bia memanfaatkan list. Jangkauan nilai quantiles adalah 0 hingga 1 (ini merepresentasikan persentase). 

```python
plt.violinplot(data,
               showextrema=False,  
               showmeans=False,
               showmedians=False,
               quantiles=None)

# set label title, sumbu x, dan sumbu y.
plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')


plt.show()
```

#### Horisontal Violin Plot

Untuk menghasilkan violin plot dalam bentuk horisontal, kita hanya perlu menambahkan parameter vert (vertikal) dengan nilai False pada fungsi `plt.boxplot`.

```python
plt.violinplot(data, vert=False)

# set label title, sumbu x, dan sumbu y.
plt.title('Horisontal Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')


plt.show()
```

#### Multiple Violin Plot

Cara-cara membuat multiple violin plot adalah:
- Membuat beberapa variabel yang berisi suatu data, misal sekumpulan angka acak.
- Menyatukan variabel tersebut ke dalam list.
- Menggunakan fungsi `plt.violinplot` dengan parameter data yang akan di plotting.

```python
np.random.seed(10)

# 4 variabel ini akan menampung sekumpulan bilangan acak.
# bilangan acak tersebut akan membentuk distribusi normal.
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)

#menyatukan 4 variabel ke dalam list.
data = [data1, data2, data3, data4]
```

```python
plt.violinplot(data)

# set label title, sumbu x, dan sumbu y.
plt.title('Multiple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')


plt.show()
```
