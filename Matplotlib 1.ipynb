#### Membuat Suatu Plotting Sederhana

Terdapat 2 cara untuk melakukan plotting atau visualisasi data.

- Membuat suatu figure yang memiliki sebuah axes dan melakukan plotting data pada axes.

```python
#Pertama, kita harus menyiapkan data sample.
#Pada contoh ini, data samplenya terdiri dari x dan y.
#Nilai x dan y akan kita visualisasikan atau kita plotting.
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

#Kedua, kita akan membuat suatu object figure dan sebuah object axes dengan fungsi seperti dibawah ini.
#Kita harus menyiapkan 2 variabel untuk menampung figure dan axes.
#Variabel fig akan menampung object figure dan variabel ax akan menampung object axes.
fig, ax = plt.subplots()

#Ketiga, kita akan melakukan visualisasi data atau plotting x dan y dengan method plot yang ada dalam object axes.
ax.plot(x, y)
```

- Untuk melakukan plotting data, kita bisa menggunakan alternatif lain yaitu dengan langsung memanfaatkan method plot() pada 
pyplot untuk melakukan plotting sederhana.

```python
#Pertama, kita menyiapkan data sample x dan y.
#Data sample ini sama dengan contoh sebelumnya.
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

#Disini kita tidak membuat object figure dan axes. Kita langsung memanfaatkan method plot() pada pyplot.
#Dalam contoh ini kita akan memplotting x dan y.
plt.plot(x,y)

#Dari output, kita bisa lihat bahwa visualisasi data dengan cara ini akan menghasilkan hasil yang sama dengan cara sebelumnya.
```

#### Komponen dari Sebuah Matplotlib Figure

Gambar anatomy of a figure sangatlah penting untuk kita pahami. Tetapi gambar ini tidak bisa langsung kita pahami begitu saja.
Kita akan memahaminya secara bertahap. Untuk sekarang sebaiknya kita pahami terlebih dahulu apa itu figure, axes dan axis sehingga kita bisa membedakannya.


- Figure adalah media dasar atau container dasar untuk melakukan visualisasi data dengan menggunakan matplotlib. Dengan kata lain, figure bisa kita sebut sebagai container besarnya. 
- Dalam sebuah figure, terdapat satu atau lebih axes.
- Axes adalah area gambarnya atau area untuk melakukan plotting. Axes akan selalu berada di dalam sebuah figure.
- Axis adalah indikasi dari sumbu (x dan y). Jumlah axis tergantung pada visualisasi data. 
- Untuk visualisasi data 2 dimensi, jumlah axisnya 2 (x dan y).
- Untuk visualisasi data 3 dimensi,  jumlah axisnya 3 (x, y, dan z).

Ada beberapa cara atau method untuk membuat figure dan axes, yaitu:

```python
#Ini merupakan method jika kita ingin membuat figure tanpa adanya axes.
#Jika kita membuat figure tanpa axes, maka kita tidak bisa melakukan visualisasi data karena kita melakukan
#plotting data bukan langsung dari object figure, melainkan dari object axes.
fig = plt.figure()
```

```python
#Ini merupakan method jika kita ingin membuat figure dengan sebuah axes.
#Kita harus menyiapkan 2 variabel untuk menampung object figure dan object axes.
#Dalam contoh ini, variabel fig akan menampung object figure, sedangkan 
#variabel ax akan menampung object axes.
fig, ax = plt.subplots()
```

```python
#Ini merupakan method jika kita ingin membuat figure dengan sebuah axes dan menyertakan parameternya (a, b).
#Contoh dibawah ini akan membuat figure dengan 2x3 grid axes.
#Artinya kita akan membuat object figure dan 6 buah object axes.
#Variabel fig akan menampung object figure 
#Variabel axs akan menampung 6 buah object axes.
fig, axs = plt.subplots(2, 3)

#Untuk pengertian parameternya adalah a merupakan jumlah baris, dan b merupakan jumlah kolom. 
#Dalam contoh dibawah artinya akan ada 2 baris dan 3 kolom.
```

#### Dua Cara untuk Menggunakan Matplotlib

- OO (Object Oriented Style) untuk melakukan plotting. Ciri utama dari OO (Object Oriented Style) adalah secara eksplisit membuat figures dan axes, serta memanggil methods dari keduanya.

```python
#Pertama, kita harus mempersiapkan data frame.
#Method 'np.linspace' bisa digunakan untuk menghasilkan sejumlah deret bilangan.
#Deret bilangan ini akan ditentukan dengan parameter (a, b, c).
#Parameter a adalah starting pointnya, di contoh ini a = 0.
#Parameter b adalah end pointnya atau nilai awal, di contoh ini a = 2.
#Parameter c artinya berapa data point yang ingin kita bentuk.
#Di contoh ini c = 100, artinya adalah kita ingin membentuk 100 data point dengan nilai dari 0 sampai dengan 2.
x = np.linspace(0, 2, 100)
x
```

```python
#Kedua, kita akan memvisualisasikan data dari data sample sebelumnya.
#Caranya adalah kita harus membuat sebuah figure dan sebuah axes.
#Variabel fig akan menampung object figure.
#Variabel ax akan menampung object axes.
fig, ax = plt.subplots()

#Ketiga, kita akan melakukan plotting data pada object axes.
#Dalam contoh ini kita akan melakukan plotting 3 variant data pada object axes.
#Operator '*' menandakan adanya pemangkatan, contohnya untuk nilai y pada label quadratic adalah hasil pangkat 2 dari x.
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

ax.set_xlabel('x label') #Method untuk menyertakan x-label pada axes. Contoh nama label pada contoh ini adalah 'x label'.
ax.set_ylabel('y label') #Method untuk menyertakan y-label pada axes. Contoh nama label pada contoh ini adalah 'y label'.
ax.set_title('Simple Plot') #Method ini untuk menyertakan title atau judul pada axes. Contohnya adalah 'Simple Plot'.
ax.legend() #Method ini untuk menyertakan legend pada axes.
```

- Pyplot Style untuk Melakukan Plotting
Ciri utama dari pyplot style adalah kita akan mengandalkan pyplot untuk membuat dan mengelola figures dan axes, serta melakukan plotting dengan menggunakan fungsi pyplot. Kita tidak perlu membentuk figure dan axes secara eksplisit.

```python
#Pertama, kita harus mempersiapkan data frame.
#Data frame ini sama dengan contoh sebelumnya,
#dimana kita memanfaatkan method'np.linspace' untuk menghasilkan sejumlah deret bilangan.
x = np.linspace(0, 2, 100)
x
```

```python
#Kedua kita bisa langsung melakukan plotting data dengan memanfaatkan 'plt.plot'.
#Disini kita tidak harus membentuk object figure dan axes.
#Perbedaannya dengan plotting sebelumnya adalah pemanggilan method plotnya bukan pada object axes,
#melainkan pada plt.
#Dalam contoh ini kita akan melakukan plotting 3 variant data pada axes.
#Operator '*' menandakan adanya pemangkatan, contohnya untuk nilai y pada label quadratic adalah hasil pangkat 2 dari x.
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

#Perbedaannya tahap ini dari tahap OO (Object Oriented Style) adalah kita cukup memanggil method seperti dibawah ini.
plt.xlabel('x label') #Method untuk menyertakan x-label pada axes. Contoh nama label pada contoh ini adalah 'x label'.
plt.ylabel('y label') #Method untuk menyertakan y-label pada axes. Contoh nama label pada contoh ini adalah 'y label'.
plt.title('Simple Plot') #Method ini untuk menyertakan title atau judul pada axes. Contohnya adalah 'Simple Plot'.
plt.legend() #Method ini untuk menyertakan legend pada axes.

#Hasil plotting akan sama dengan hasil plotting menggunakan cara OO (Object Oriented Style).
```
