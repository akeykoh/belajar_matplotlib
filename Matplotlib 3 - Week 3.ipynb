#### Simple Line Plot

Langkah-langkah simple line plot:
- Menyiapkan sample data set 1 yang akan ditampung dalam variabel x. Dalam contoh kali ini kita akan mengenerate data dengan method `np.arange`.
- Menyiapkan sample data set 2 yang akan ditampung dalam variabel s. Data set 2 akan menampung nilai hasil operasi sin dari data set yang ditampung dalam variabel x. Kita bisa memanfaatkan method `np.sin`. Tetapi nilai data x tidak langsung dikenakan sin, melainkan akan dikenakan operasi perkalian terlebih dahulu `(2 * np.pi * x)`.
- Melakukan plotting untuk dua data set yang kita punya. Sumbu x akan merepresentasikan nilai yang ditampung variabel s dan sumbu y merepresentasikan sinus.
- Memberi label untuk sumbu x, sumbu y, dan title atau judul dengan method `.set`.

```python
#Method "np.arange" akan menyertakan 3 parameter yaitu (0.0, 2.0, 0.01).
#Parameter 1 merepresentasikan nilai starting point dan bersifat inklusif (akan disertakan dalam deret bilangan).
#Parameter 2 merepresentasikan nilai akhir dan bersifat eksklusif (hanya digunakan sebagai pembatas saja).
#Parameter 3 berfungsi untuk menentukan interval nilainya. 
x = np.arange(0.0, 2.0, 0.01)
x
```

```python
#(2 * np.pi * x) mengindikasikan bahwa nilai x adalah nilai radian.
s = np.sin(2 * np.pi * x)
s
```

```python
#method plt.subplots akan menghasilkan dua buah nilai yang akan ditampung dalam dua buah variabel.
#variabel pertama akan menampung figure.
#variabel kedua akan menampung axes. 
fig, ax = plt.subplots() # 00 Style (Object Oriented Style)

#Melakukan plotting pada axes.
ax.plot(x, s)

#Nilai parameter berupa String.
ax.set(xlabel='nilai x',
       ylabel='nilai sin(x)',
       title='Visualisasi nilai sin')

#Kita bisa menampilkan grid dengan method ax.grid().
ax.grid()

plt.show()
```

```python
#Membuat plotting sederhana dengan memanfaatkan method ".plot" (pyplot style)
plt.plot(x, s)

#Set label sumbu x, sumbu y, dan title.
plt.xlabel('nilai x')
plt.ylabel('nilai sin(x)')
plt.title('Visualisasi nilai sin')

#Menampilkan grid dengan method ax.grid()
plt.grid()

plt.show()
```

#### Multiple Subplots

Untuk mempelajari multiple suplots, disini kita akan menyediakan 4 deret bilangan yang ditampung dalam 4 variabel.

```python
#Nilai x1 dan x2 menggunakan method np.linspace. 
#Parameter 1 merepresentasikan nilai minimum dan bersifat inklusif.
#Parameter 2 merepresentasikan nilai maksimum dan bersifat inklusif.
#Parameter 3 mengindikasikan jumlah data poin yang akan di generate. 
x1 = np.linspace(0.0, 5.0, 100)
x2 = np.linspace(0.0, 2.0, 100)

#Nilai y1 dan y2 menggunakan method np.cos.
#Variabel y1 dan y2 akan menampung deret bilangan dari hasil operasi cos variabel x1 dan x2. 
y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
#Untuk melihat nilai-nilai x1
x1
```

#### Multiple Subplots dengan OO Style

Kita dapat membuat multiple subplots dengan method `plt.suplots`. Disini kita akan membuat multiple subplots dengan 2 baris dan 1 kolom. 

```python
#Parameter 1 merepresentasikan baris.
#Parameter 2 merepresentasikan kolom.
#Kita akan membuat 2 variabel untuk menampung hasilnya.
#Variabel pertama adalah fig untuk menampung object figure.
#Variabel kedua adalah berupa tuple (ax1, ax2) yang menampung axes 1 dan axes 2.
fig, (ax1, ax2) = plt.subplots(2,1)

#Kita bisa mengatur judul pada figure dengan method .suptitle.
fig.suptitle('Multiple Subplots')

#Kita akan melakukan plotting x1 dan y1
#Parameter ketiga akan mengatur warna dan bentuk marker.
ax1.plot(x1, y1, 'ro-')

#Memberikan label dengan method set. 
#Kita bisa merapikan label dengan membuat x1 menjadi subscript menggunakan latex style
#Caranya dengan mengapit bagian yang mau diatur dengan latex menggunakan simbol $.
ax1.set_ylabel('Nilai $cos(x_1)$')

#Kita akan melakukan plotting x2 dan y2
#Parameter ketiga akan mengatur warna dan bentuk marker.
ax2.plot(x2, y2, 'g.-')

#Memberikan label dengan method set. 
ax2.set_ylabel('Nilai $cos(x2)$')

#Memberikan label dengan method set. 
#Nilai label sumbu x hanya diterapkan di axes kedua saja.
ax2.set_xlabel('Nilai x')

plt.show()
```
