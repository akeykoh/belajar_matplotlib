#### Cara Import Module dan Mengecek Version

Kegunaannya adalah untuk memastikan setiap output dari matplotlib akan ditampilkan dalam halaman notebook yang sama. Sedangkan untuk Jupyter Notebook versi lebih baru, tidak diperlukan karena secara default jika kita melakukan plotting dengan matplotlib, maka hasilnya akan secara otomatis ditampilkan pada notebook.

```python
# Ini merupakan cara untuk mengimport module.
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

# Ini merupakan cara mengecek version module.
print(matplotlib.__version__)
print(pd.__version__)
```

#### Sample Data Set

Disini kita akan membentuk Pandas data framenya dari struktur data Dictionary yang terdiri dari 6 kolom yaitu kolom nama, usia, gender, provinsi, n_anak, dan n_pets.

```python
df = pd.DataFrame({
    'nama':['cecep', 'karti', 'bejo', 'tejo', 'asep', 'wati', 'amat'],
    'usia':[23, 78, 22, 19, 45, 33, 20],
    'gender':['L', 'P', 'L', 'L', 'L', 'P', 'L'],
    'provinsi':['jabar', 'jatim', 'jabar', 'jatim', 'jabar', 'jateng', 'jateng'],
    'n_anak':[2, 0, 0, 3, 2, 1, 4],
    'n_pets':[5, 1, 0, 5, 2, 2, 3]
})

df
```

#### Scatter Plot

Kita bisa melakukan plotting data dengan memanfaatkan method `df.plot()` dan menyertakan beberapa parameter. Parameter `kind` akan menentukan jenis plotting. Parameter `x` diisi dengan nama kolom yang mau dijadikan acuan. Parameter `y` diisi dengan nama kolom yang mau dijadikan acuan. Parameter color untuk menentukan warna dari marker.

```python
# melakukan plotting
df.plot(kind='scatter', x='n_anak', y='n_pets', color = 'red')

# mengatur label
plt.title('Scatter Plot')
plt.xlabel('Jumlah Anak')
plt.ylabel('Jumlah Pets')
plt.show()
```

#### Histogram

Dibawah ini kita akan membuat plotting histogram dari kolom usia dengan method `df[['usia']].plot` dan menyertakan beberapa parameter. Parameter `kind` akan menentukan jenis plotting. Di contoh ini, parameter `bins` diisi dengan suatu List yang berisi beberapa bilangan. 

```python
# melakukan plotting
df[['usia']].plot(kind='hist', bins=[0, 20, 40, 60, 80, 100])

# mengatur label
plt.title('Histogram')
plt.xlabel('Usia')
plt.ylabel('Jumlah')
plt.show()
```

#### Line Plot

Kita bisa melakukan plotting data dengan jenis Line Plot dengan memanfaatkan method `df.plot()` dan menyertakan beberapa parameter. Parameter `kind` diisi dengan nilai `line`. Parameter x diisi dengan nama kolom yang mau dijadikan acuan. Parameter y diisi dengan nama kolom yang mau dijadikan acuan. Parameter color untuk menentukan warna dari marker.

```python
# melakukan plotting
df.plot(kind='line', x='nama', y='n_anak', color ='green')

# mengatur label
plt.title('Line Plot')
plt.ylabel('Jumlah Anak')
plt.show()
```

#### Multiple Line Plot

Ketika kita memiliki multiple line plot dan kita ingin hasilnya digambarkan di 1 axes yang sama, kita bisa memanggil method `plt.gca()`. Ketika kita melakukan plottong, kita juga harus menyertakan parameter `ax` dengan nilai object axes yang baru ditangkap dari pemanggilan method `plt.gca()`.

```python
# melakukan plotting
df.plot(kind='line', x='nama', y='n_anak', color ='green') # line plot 1
df.plot(kind='line', x='nama', y='n_pets', color ='red')   # line plot 2

# mengatur label
plt.title('Line Plot')
plt.ylabel('Jumlah')
plt.show()
```

```python
ax = plt.gca() # mengembalikan object axes

# melakukan plotting
df.plot(kind='line', x='nama', y='n_anak', color ='green', ax=ax) # line plot 1
df.plot(kind='line', x='nama', y='n_pets', color ='red', ax=ax)   # line plot 2

# mengatur label
plt.title(' Multiple Line Plot')
plt.ylabel('Jumlah')
plt.show()
```

#### Bar Plot

Kita bisa melakukan plotting data dengan jenis Bar Plot dengan memanfaatkan method `df.plot()` dan menyertakan beberapa parameter. Parameter `kind` diisi dengan nilai `bar`. Parameter x diisi dengan nama kolom yang mau dijadikan acuan. Parameter y diisi dengan nama kolom yang mau dijadikan acuan.

```python
# melakukan plotting
df.plot(kind='bar', x='nama', y='usia')

# mengatur label
plt.title('Bar Plot')
plt.ylabel('Usia')
plt.show()
```

```python
# melakukan grouping data berdasarkan kolom provinsi
# kita akan menghitung kolom nama dan hasilnya akan di plotting
df.groupby('provinsi')['nama'].count().plot(kind='bar')

# mengatur label
plt.title('Bar Plot')
plt.ylabel('Jumlah')
plt.show()
```

#### Stacked Bar Plot dengan Dua Level Group By

Kita bisa melakukan grouping 2 tingkat dengan cara memanggil method `df.groupby()` dan menyertakan List berisi 2 item (item 1 akan berisi nama kolom yang digunakan untuk pengelompokkan level 1 dan item 2 akan berisi nama kolom yang digunakan untuk pengelompokkan level 2). 

```python
# melakukan grouping 2 level dan plotting.
df.groupby(['provinsi', 'gender']).size().unstack().plot(kind='bar',
                                                         stacked=True)
# mengatur label
plt.title('Scatter Bar Plot')
plt.ylabel('Jumlah')
plt.show()
```

```python
# melakukan grouping 2 level dan plotting.
df.groupby(['gender', 'provinsi']).size().unstack().plot(kind='bar',
                                                         stacked=True)
# mengatur label
plt.title('Scatter Bar Plot')
plt.ylabel('Jumlah')
plt.show()
```
