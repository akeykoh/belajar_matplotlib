#### 3D Axes

Untuk 3D axes, kita harus membangun sendiri object figure dan axesnya. Axes yang akan kita bentuk adalah axes untuk 3 dimensi, caranya adalah menyertakan parameter projection dengan nilai 3d. 

```python
fig = plt.figure() # membentuk object figure
ax = plt.axes(projection='3d') # membentuk object axes

plt.show()
```

#### 3D Line Plot

Kita akan mencoba melakukan plotting untuk 3 variabel pada area 3 dimensi. Caranya adalah:
- Tentukan 3 variabel dan isi datanya untuk di plotting.
- Membentuk object figure dengan method `plt.figure()`.
- Membentuk object 3D axes dengan method `plt.axes(projection='3d')`.
- Melakukan plotting berupa line plot dengan memanggil `ax.plot`. 
- Melakukan pengaturan label dengan `ax.set`.
- Melihat hasil 3Dline plot dengan `plt.show()`

```python
# 3 variabel yang akan di plotting pada area 3 dimensi.
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z) 
y = z * np.cos(20 * z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot(x, y, z, color='red') # color untuk mengatur warna line plot

ax.set(title='Contoh 3D Line Plot',
       xlabel = 'Sumbu X',
       ylabel = 'Sumbu Y',
       zlabel = 'Sumbu Z')

plt.show()
```

#### 3D Scatter Plot

Kita juga bisa menghasilkan scatter plot untuk 3 variabel pada area 3 dimensi. Caranya adalah:
- Tentukan 3 variabel dan isi datanya untuk di plotting.
- Membentuk object figure dengan method `plt.figure()`.
- Membentuk object 3D axes dengan method `plt.axes(projection='3d')`.
- Melakukan plotting berupa scatter plot dengan memanggil `ax.scatter`. 
- Melakukan pengaturan label dengan `ax.set`.
- Melihat hasil 3Dline plot dengan `plt.show()`

```python
# 3 variabel yang akan di plotting pada area 3 dimensi.
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z) 
y = z * np.cos(20 * z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter(x, y, z, color='red') # color untuk mengatur warna line plot

ax.set(title='Contoh 3D Scatter Plot',
       xlabel = 'Sumbu X',
       ylabel = 'Sumbu Y',
       zlabel = 'Sumbu Z')

plt.show()
```
