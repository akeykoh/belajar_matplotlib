#### Cara Import Module dan Mengecek Version

Disini kita akan mengaktifkan magic comment yaitu `%matplotlib notebook`. Sebenarnya magic comment ini mirip dengan `%matplotlib inline` yaitu untuk memastikan output yang dihasilkan matplotlib bisa tampil di halaman Jupyter Notebook. Yang berbeda adalah dengan `%matplotlib notebook`, kita bisa berinteraksi dengan 3D plotting yang dihasilkan.

```python
# %matplotlib inline
%matplotlib notebook
```

```python
#Ini merupakan cara untuk mengimport module.
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#Ini merupakan cara mengecek version module.
print(matplotlib.__version__)
print(np.__version__)
```

#### 3D Axes

Untuk 3D axes, kita harus membangun sendiri object figure dan axesnya. Axes yang akan kita bentuk adalah axes untuk 3 dimensi, caranya adalah menyertakan parameter projection dengan nilai 3d. 

```python
fig = plt.figure() # membentuk object figure
ax = plt.axes(projection='3d') # membentuk object axes

plt.show()
```
