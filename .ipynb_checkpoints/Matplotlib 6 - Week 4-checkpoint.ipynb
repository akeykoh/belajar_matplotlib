#### Simple Pie Plot

Simple pie plot digunakan untuk melakukan plotting data dimana kategori berasosiasi dengan nilai tertentu. Untuk menghasilkan simple pie plot, kita bisa menggunakan method `plt.pie`. Pemanggilan method `plt.pie` membutuhkan suatu nilai atau sekumpulan nilai dengan tipe data Integer. 

```python
kategori = ['A', 'B', 'C', 'D', 'E'] #list dari deret karakter.

#nilai dalam variabel data akan dibuat proporsi persentasenya.
data = [215, 130, 245, 210, 225] #list dari sekumpulan angka.
```

```python
plt.pie(data,
        labels=kategori,   #parameter labels bersifat opsional.
        autopct='%1.1f%%', #untuk menunjukkan persentase dari setiap area pada pyplot dan opsional juga.
        startangle=90)     #untuk melakukan rotasi pada pyplot.

plt.title('Contoh Pie Plot') #untuk mengatur judul.

plt.show()
```

#### Mengatur Legend dan Color

Kita bisa mengatur color dengan menyertakan parameter color dalam method `plt.pie`. Untuk memunculkan legend atau keterangan plotting, kita bisa menggunakan method `plt.legend()`.

```python
kategori = ['A', 'B', 'C', 'D', 'E'] #list dari deret karakter.
data = [215, 130, 245, 210, 225] #list dari sekumpulan angka.
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue'] #list dari sekumpulan data String.
```

```python
plt.pie(data,
        labels=kategori,   #parameter labels bersifat opsional.
        colors=warna,      #untuk mengatur warna.
        autopct='%1.1f%%', #untuk menunjukkan persentase dari setiap area pada pyplot dan opsional juga.
        startangle=90)     #untuk melakukan rotasi pada pyplot.

plt.legend()

plt.title('Contoh Pie Plot') #untuk mengatur judul.
    
plt.show()
```

#### Exploded Pie Plot

Untuk membuat exploded pie plot, kita bisa menyertakan parameter explode ketika melakukan plotting data dengan method `plt.pie`. Kita akan menambahkan data set explode_var, semakin besar nilai explode_varnya maka akan semakin jauh juga pergeseran yang dilakukan terhadap potongan dari pyplot. 

```python
kategori = ['A', 'B', 'C', 'D', 'E'] #list dari deret karakter.
data = [215, 130, 245, 210, 225] #list dari sekumpulan angka.
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue'] #list dari sekumpulan data String.
explode_var = [0., 0., 0.2, 0., 0.08] #list plotting point atau pecahan dan akan berkolerasi dengan kategori.
#Contohnya, kategori C dan E akan menjauh sedikit dari pyplot kita karena nilai explode_varnya 0.2 dan 0.08.
```

```python
plt.pie(data,
        labels=kategori,     #parameter labels bersifat opsional.
        colors=warna,        #untuk mengatur warna.
        explode=explode_var, #parameter explode akan berkolerasi dengan variabel explode_var.
        autopct='%1.1f%%',   #untuk menunjukkan persentase dari setiap area pada pyplot dan opsional juga.
        startangle=90)       #untuk melakukan rotasi pada pyplot.

plt.legend()
plt.title('Contoh Exploded Pie Plot') #untuk mengatur judul.

plt.show()
```
