#### Cara Import Module dan Mengecek Version

Sebelum mengimport module, kita bisa menjalankan magic comment. `%matplotlib inline` dikenal sebagai magic comment pada Jupyter Notebook dan Jupyter Lab. Kegunaannya adalah untuk memastikan setiap output dari matplotlib akan ditampilkan dalam halaman notebook yang sama. Magic comment ini harus kita eksekusi, terutama jika kita menggunakan Jupyter Notebook versi lama. Sedangkan untuk Jupyter Notebook versi lebih baru, tidak diperlukan karena secara default jika kita melakukan plotting dengan matplotlib, maka hasilnya akan secara otomatis ditampilkan pada notebook.

```python
%matplotlib inline
```

```python
#Ini merupakan cara untuk mengimport module.
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#Ini merupakan cara mengecek version module.
print(matplotlib.__version__)
print(np.__version__)
```

#### Sample Dataset

Kita akan menyiapkan sample dataset yang terdiri dari deret bilangan dengan nilai min -3 dan nilai max 3, total data point yang dihasilkan adalah 100 data point. Untuk membentuk data surface / area, kita bisa menggunakan method `np.meshgrid()`.

```python
n = np.linspace(-3.0, 3.0, 100)

x,y = np.meshgrid(n, n)  # menghasilkan data 2 dimensi yang ditampung pada variabel x dan y.
z = np.sqrt(x**2 + y**2) # menampung hasil akar kuadrat dari penjumlahan x kuadran dan y kuadrat.

print(x.shape)
print(y.shape)
print(z.shape)
```

#### Contour Plot

Contour plot bisa digunakan untuk memvisualisasikan data 3 dimensi dalam format 2 dimensi. Kita sering menemui contour plot ketika berinteraksi dengan data geospasial.

```python
plt.contour(x, y, z) # 3 parameter merepresentasikan koordinat x, y, dan z.
plt.colorbar()       # matriks untuk membantu kita menginterpretasikan semua warna dalam bentu angka.

plt.title('Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

#### Filled Contour Plot

Countour plot hanya menampilkan garisnya saja, tetapi filled contour plot akan memenuhi areanya dengan warna.

```python
plt.contourf(x, y, z)
plt.colorbar()

plt.title('Filled Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
