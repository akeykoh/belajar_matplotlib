#### Simple Scatter Plot

Disini kita akan membentuk sactter plot sederhana dengan 2 data set yaitu x dan y. Untuk membentuk scatter plot dengan Matplotlib, kita bisa memanggil method `plt.scatter` diikuti dengan parameter. Disini kita memasukkan parameter data setnya (x dan y, ini wajib disertakan), parameter label (opsional), dan parameter color (opsional).

```python
x = [2,4,6,8,10,11,11.5,11.7] #list yang menampung sekumpulan bilangan integer dan float.
y = [1,1.5,2,2.5,3,3.5,4,4.5] #list yang menampung sekumpulan bilangan integer dan float.
```

```python
plt.scatter(x, y, label='Data 1', color='r')

#Set label sumbu x, sumbu y, dan title.
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Scatter Plot')
#Untuk menampilkan legend.
plt.legend()

plt.show()
```

#### Multiple Scatter Plot

Untuk membentuk scatter plot dari data point 1 dan 2, langkah-langkahnya adalah:
- Melakukan plotting data point 1 dengan method `plt.scatter` dan menyertakan parameter x1 dan y1, ini wajib disertakan), parameter label (opsional), dan parameter color (opsional).
- Melakukan plotting data point 2 dengan method `plt.scatter` dan menyertakan parameter x2 dan y2, ini wajib disertakan), parameter label (opsional), dan parameter color (opsional).
- Memberikan label untuk sumbu x, sumbu y, dan title.
- Menampilkan legend dengan method `plt.legend()`.
- Melihat hasil scatter plot dengan method `plt.show()`.

```python
#Akan membentuk kumpulan data point 1
x1 = [2,4,6,8,10,11,11.5,11.7] #list yang menampung sekumpulan bilangan integer dan float.
y1 = [1,1.5,2,2.5,3,3.5,4,4.5] #list yang menampung sekumpulan bilangan integer dan float.

#Akan membentuk kumpulan data point 2
x2 = [8,8.5,9, 9.5,10,10.5,11] #list yang menampung sekumpulan bilangan integer dan float.
y2 = [3,3.5,3.7,4,4.5,5,5.2]   #list yang menampung sekumpulan bilangan integer dan float.
```

```python
plt.scatter(x1, y1, label='Data 1', color='r')
plt.scatter(x2, y2, label='Data 2', color='b')

#Set label sumbu x, sumbu y, dan title.
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Multiple Scatter Plot')
#Untuk menampilkan legend.
plt.legend()

plt.show()
```

#### Pengaturan Marker

Disini kita akan mengatur marker dalam scatter plot kita dengan cara menambah parameter pada method `plt.scatter`. Pada contoh ini kita menambah parameter color (untuk mengatur warna), parameter linewidths (mengatur ketebalan line), parameter marker (untuk mengatur bentuk marker), parameter edge color (untuk mengatur warna garis tepi), dan parameter s (untuk mengatur ukuran marker).

```python
#Akan membentuk kumpulan data point 1
x1 = [2,4,6,8,10,11,11.5,11.7] #list yang menampung sekumpulan bilangan integer dan float.
y1 = [1,1.5,2,2.5,3,3.5,4,4.5] #list yang menampung sekumpulan bilangan integer dan float.

#Akan membentuk kumpulan data point 2
x2 = [8,8.5,9, 9.5,10,10.5,11] #list yang menampung sekumpulan bilangan integer dan float.
y2 = [3,3.5,3.7,4,4.5,5,5.2]   #list yang menampung sekumpulan bilangan integer dan float.
```

```python
plt.scatter(x1, y1,
           color='cyan',
           linewidths=1,
           marker='s',
           edgecolor='red',
           s=100)

plt.scatter(x2, y2,
           color='yellow',
           linewidths=1,
           marker='^',
           edgecolor='blue',
           s=200)

plt.show()
```

#### Scatter Plot untuk Data Multi Dimensi

Disini kita akan mengimpor suatu dataset yaitu `load_iris` dari modul `sk.learn`. Modul ini biasanya digunakan ketika kita belajar machine learning. `load_iris` berguna untuk mengembalikan suatu data yang mempunyai struktur data Dictionary.

```python
from sklearn.datasets import load_iris

#Menampung load_iris ke dalam variabel iris.
iris = load_iris()
features = iris['data'].T
```

```python
#Melihat key yang ada dalam dictionary iris. 
iris.keys()
```

```python
#Melihat isi dari dictionary iris dengan memanggil salah satu keynya yaitu data.
iris['data']
```

```python
#Melihat nama feature.
iris['feature_names']
```

```python
#Mengelompokkan data per feature. 
#T digunakan untuk melakukan proses transpose (membalik kolom dan baris).
features = iris['data'].T
features
```

```python
#Melihat nilai dalam target.
iris['target']
```

```python
#Melihat nama target.
iris['target_names']
```

```python
#Melakukan plotting data dari data set iris.
plt.scatter(features[0], features[1], #melakukan plotting terkai sepal length dan sepal width.
            alpha = 0.2,              #mengatur transparansi plotting
            s=100 * features[3],      #mengatur ukuran marker
            c=iris['target'],         #nilai yang mau di mapping ke dalam warna akan ditempatkan ke parameter c.
            cmap='viridis')           #color map digunakan untuk mekanisme mapping terhadap sekumpulan bilangan tertentu dengan warna.


plt.xlabel(iris['feature_names'][0])
plt.ylabel(iris['feature_names'][1])
plt.title('Iris Dataset')

#Untuk menampilkan color bar.
plt.colorbar()

plt.show()
```
