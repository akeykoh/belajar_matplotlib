#### Simple Bar Plot

Kita akan memvisualisasikan 2 data set menjadi bar plot dengan menggunakan method `plt.bar`. 2 data set tersebut adalah data dan kategori. Terdapat 2 parameter dalam method `plt.bar`, biasanya parameter pertama bersifat categorical dan parameter dua berisi sederet nilai yang akan berasosiasi dengan parameter pertama. 

Perbedaan histogram dan bar plot adalah histogram digunakan untuk melakukan plotting data numerik dan memiliki sifat continuous, sedangkan bar plot biasanya digunakan untuk melakukan plotting data yang memiliki sifat categorical / discrete.

```python
data = [25, 45, 55, 125, 225] #list dari sekumpulan angka.
kategori = ['A', 'B', 'C', 'D', 'E'] #list dari deret karakter.
```

```python
plt.bar(kategori, data)


#Kita bisa menampilkan grid dengan method plt.grid().
plt.grid()

#Set label sumbu x, sumbu y, dan title.
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

#### Mengatur Grid dan Color

Kita dapat mengatur grid dan color pada bar plot. 

Untuk mengatur warna, bisa kita lakukan ketika memanggil method `plt.bar` dan menyertakan parameter color (mengatur warna) dan alpha (mengatur transparasi warna). 

Untuk mengatur grid, bisa kita lakukan ketika memanggil method `plt.grid` dan menyertakan parameter linetyle (menentukan bentuk line dalam bar plot), linewidth (menentukan ketebalan line dalam bar plot), axix (menentukan linenya berasosiasi dengan nilai sumbu apa), dan alpha (mengatur transparasi line).

```python
data = [25, 45, 55, 125, 225] #list dari sekumpulan angka.
kategori = ['A', 'B', 'C', 'D', 'E'] #list dari deret karakter.
```

```python
plt.bar(kategori,
        data,
        color='red', #mengatur warna.
        alpha=0.75)  #mengatur transparasi warna.

#Kita bisa menampilkan dan mengatur grid dengan method plt.grid().
plt.grid(linestyle='--', #menentukan bentuk line dalam bar plot.
         linewidth=2,    #menentukan ketebalan line dalam bar plot.
         axis='x',       #menentukan linenya berasosiasi dengan nilai sumbu apa, pilihan: x, y, both.
         alpha=0.50)     #mengatur transparasi line

#Set label sumbu x, sumbu y, dan title.
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

#### Horizontal Bar Plot

Untuk menghasilkan horizontal bar plot, kita akan menggunakan method `plt.barh`. Untuk parameternya sama dengan contoh sebelumnya yaitu terdapat 2 parameter, biasanya parameter pertama bersifat categorical dan parameter dua berisi sederet nilai yang akan berasosiasi dengan parameter pertama. 

```python
data = [25, 45, 55, 125, 225] #list dari sekumpulan angka.
kategori = ['A', 'B', 'C', 'D', 'E'] #list dari deret karakter.
```

```python
#plt.barh(kategori, data)
plt.barh(kategori[::-1], data[::-1]) #ini bisa digunakan jika kita ingin membalikkan urutan list.

#Kita bisa menampilkan dan mengatur grid dengan method plt.grid().
plt.grid(linestyle='--',
         linewidth=1,
         axis='x', 
         alpha=0.75)

#Set label sumbu x, sumbu y, dan title.
plt.xlabel('Jumlah')
plt.ylabel('Kategori')
plt.title('Contoh Horizontal Bar Plot')

plt.show()
```

#### Stacked Bar Plot

Untuk membuat stacked bar plot, kita akan menggunakan 3 data set yaitu data1, data2, dan kategori. Stacked bar plot biasanya digunakan untuk melakukan plotting data untuk 2 data yang berbeda, tetapi memiliki kesamaan dalam kategori. Langkah-langkah membuat stacked bar plot:

- Melakukan plotting data pertama dengan method `plt.bar` dan menambahkan parameter. Yang berbeda adalah disini kita akan menambahkan parameter ketiga yaitu label.
- Melakukan plotting data kedua dengan method `plt.bar` dan menambahkan parameter. Yang berbeda adalah disini kita akan menambahkan parameter ketiga yaitu label. Kita juga harus menambahkan parameter bottom untuk menentukan data2 akan ditempatkan diatas data yang mana. Kita bebas menentukan dimana menambahkan parameter bottom, di data 1 atau data 2.
- Memanggil method `plt.legend`.
- Melakukan pengaturan grid dengan method `plt.grid`.
- Melakukan pengaturan xlabel, ylabel, dan title.
- Menunjukkan stacked bar plot dengan method `plt.show()`.

```python
data1 = [25, 85, 75, 40, 60] #list dari sekumpulan angka.
data2 = [40, 35, 20, 55, 10] #list dari sekumpulan angka.
kategori = ['A', 'B', 'C', 'D', 'E'] #list dari deret karakter.
```

```python
plt.bar(kategori, data1, label='Data 1')
#plt.bar(kategori, data2, label='Data 2') #jika tidak menggunakan bottom, akan terjadi penimpahan plotting.
plt.bar(kategori, data2, label='Data 2', bottom=data1)

plt.legend() #parameter label akan berasosiasi dengan legend.

#Kita bisa menampilkan dan mengatur grid dengan method plt.grid().
plt.grid(linestyle='--',
         linewidth=1,
         axis='y', 
         alpha=0.75)

#Set label sumbu x, sumbu y, dan title.
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

#### Grouped Bar Plot

Grouped bar plot juga biasanya digunakan untuk melakukan plotting data yang memiliki 2 atau lebih data set yang berbeda dengan pembagian kategori yang sama. Jika dalam stacked bar plot, barplotnya akan ditumpuk. Sedangkan dalam grouped bar plot, bar plotnya akan dikelompokkan berdasarkan kategori dan akan di plot berdampingan satu dengan yang lain. Langkah-langkah membuat grouped bar plot:

- Menentukan width / lebar.
- Melakukan plotting dengan method `plt.bar`. Yang berbeda adalah nilai parameter pertamanya akan berasosiasi dengan index dalam variabel x. Index juga akan menjadi acuan atau titik tengahnya. Parameter keduanya adalah data setnya, setelah itu kita juga harus menyertakan parameter width dan label.
- Memanggil method `plt.xticks` untuk mengganti nilai index misalnya kita ingin mengubah nilai index sumbu x menjadi kategori.
- Melakukan pengaturan grid dengan method `plt.grid`.
- Melakukan pengaturan xlabel, ylabel, dan title.
- Memanggil method `plt.legend`.
- Menunjukkan stacked bar plot dengan method `plt.show()`.

```python
data1 = [25, 85, 75, 40, 60] #list dari sekumpulan angka.
data2 = [40, 35, 20, 55, 10] #list dari sekumpulan angka.
kategori = ['A', 'B', 'C', 'D', 'E'] #list dari deret karakter.

#variabel x akan menampung array yang dihasilkan method np.arange dengan parameter panjang dari kategori yaitu 5.
#np.arange akan menghasilkan array yang mempunyai deret bilangan.
x = np.arange(len(kategori))
x
```

```python
width = 0.35

plt.bar(x-width/2, data1, width, label='Data 1')
plt.bar(x+width/2, data2, width, label='Data 2')

plt.xticks(x, kategori)

#Kita bisa menampilkan dan mengatur grid dengan method plt.grid().
plt.grid(linestyle='--',
         linewidth=1,
         axis='y', 
         alpha=0.75)

#Set label sumbu x, sumbu y, dan title.
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Grouped Bar Plot')

plt.legend()
plt.show()
```
