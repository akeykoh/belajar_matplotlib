#### Sample Dataset

Kita akan menyiapkan sample dataset yang terdiri dari deret bilangan dengan nilai min -3 dan nilai max 3, total data point yang dihasilkan adalah 100 data point. Untuk membentuk data surface / area, kita bisa menggunakan method `np.meshgrid()`.

```python
n = np.linspace(-3.0, 3.0, 100)

x,y = np.meshgrid(n, n)  # menghasilkan data 2 dimensi yang ditampung pada variabel x dan y.
z = np.sqrt(x**2 + y**2) # menampung hasil akar kuadrat dari penjumlahan x kuadran dan y kuadrat.

print(x.shape)
print(y.shape)
print(z.shape)
```

#### 3D Wireframe

Langkah-langkah untuk menghasilkan 3D wireframe:
- Membentuk object figure dengan method `plt.figure()`.
- Membentuk object 3D axes dengan method `plt.axes(projection='3d')`.
- Melakukan plotting data dengan method `.plot_wireframe`. Method ini membutuhkan setidaknya 3 parameter. Parameter 1 berkolerasi dengan nilai koordinat x. Parameter 2 berkolerasi dengan nilai koordinat y. Parameter 3 berkolerasi dengan nilai koordinat z. Parameter 4 yaitu color (optional) untuk mengatur warna hasil plotting.
- Melakukan pengaturan label dengan `ax.set`.
- Melihat hasil 3Dline plot dengan `plt.show()`

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_wireframe(x, y, z, color='orange') # color untuk mengatur warna line plot

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Wireframe')

plt.show()
```
